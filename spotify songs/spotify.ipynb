```python
#analyzing the genre with the most entries
genre = data['playlist_genre'].value_counts()
genre
```

HYPOTHESIS 1: POP IS THE MOST POPULAR GENRE

```python
#creating a list to use as an dataframe later
output = []
for x in genre.index:
    genpop = data.loc[data.playlist_genre == x].track_popularity.mean()
    output.append([x,genpop])
```

```python
#testing if the list works
output
```

```python
#creating a new dataframe
genre_popularity_df = pd.DataFrame(output, columns=['Genre', 'average_popularity'])
genre_popularity_df
```

```python
#change the scale in average_popularity to 0-1
genre_popularity_df.average_popularity = genre_popularity_df.average_popularity/genre_popularity_df.average_popularity.max()
genre_popularity_df
```

```python
genre_popularity_df.plot(x='Genre', kind='bar', legend= False, ylabel='Popularity from 0-1')
plt.title('Most popular Genres')
plt.xticks(rotation=45, ha='right')
plt.show()
```

As we can see: The most popular genre is actually pop music followed by latin music. The least popular genre is edm music.

HYPOTHESIS 2: TRACK POPULARITY CORRELATES TO DANCEABILITY

```python
# only selecting the numeric values
numvalues = data.select_dtypes(include=['float64', 'int64'])
numvalues
```

```python
# save it in a variable
numvalues_corr = numvalues.corr()
```

```python
#setting the width and height for the plot
plt.figure(figsize=(10,8))
#creating a heatmap to check the correlations between attributes
sns.heatmap(round(numvalues_corr, 3), annot=True)
plt.xticks(rotation=45, ha='right')
plt.title('Correlation between song attributes')
plt.show()
```

This hypothesis is wrong since danceability has nothing to do with the popularity of a song. This plot also shows that the other variables do not have much in common with the popularity since the values are low. The highest correlation for track popularity is the acousticness.
